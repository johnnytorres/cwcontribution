```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
ds = pd.read_csv('../dataset/enwiki-discussions/rafael_correa_opinions_subjectivity.csv')
ds.shape
```

```python
ds['creation_dt'] = pd.to_datetime(ds['timestamp'])
```

```python
sentiment = ds.groupby(['subjectivity']).agg({'subjectivity':'count'})
sentiment.head()
```

```python
sentiment.plot.barh()
```

```python
subjectivity = ds[ds.subjectivity=='obj']
subjectivity = subjectivity.groupby(['creation_dt']).agg({'subjectivity':'count'})
subjectivity.head()
```

```python
objectivity = ds[ds.subjectivity=='subj']
objectivity = objectivity.groupby(['creation_dt']).agg({'subjectivity':'count'})
objectivity.head()
```

```python
plt.plot(subjectivity, c='r')
#plt.hold(True)
plt.plot(objectivity, c='g')
#plt.yscale('log')
```
