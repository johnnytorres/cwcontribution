```python
import pandas as pd
```

```python
ds = pd.read_csv('../dataset/wiki/aawd.csv')
```

```python
ds.loc[:, 'stance'] = 0
ds.loc[ds.positive.isnull(), 'positive'] = False
ds.loc[ds.negative.isnull(), 'negative'] = False
ds.loc[ds.positive & ~ds.negative, 'stance'] = 1
ds.loc[ds.negative & ~ds.positive, 'stance'] = -1
ds['type'] = ds.positive_types
ds['text'] = ds.transcript
ds = ds[~ds.stance.isnull()]
ds = ds[~ds.text.isnull()]
```

```python
types = ds.groupby('type').size()
types.plot.barh()
```

```python
stances = ds.groupby('stance').size()
stances#.plot.barh()
```
