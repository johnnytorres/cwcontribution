```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import os
import folium 
from folium.plugins import MarkerCluster
```

```python
geo_pages = pd.read_csv('../dataset/geo_pages_ec.csv')
geo_pages.shape
```

```python
geo_pages.head()
```

```python
pages_coor = geo_pages[['page.page_title', 'geo_tags.gt_lat', 'geo_tags.gt_lon']].as_matrix()
pages_coor.shape
```

```python
m = folium.Map(location=[np.mean(pages_coor[:,1]), np.mean(pages_coor[:,2])], tiles='Cartodb Positron', zoom_start=1)
m.add_child(MarkerCluster(locations=pages_coor[:,1:3], popups=pages_coor[:,0]))
m.save('../results/geo_pages.html')
m
```
