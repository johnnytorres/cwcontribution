```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = pd.read_csv('../dataset/enwiki/opinions_arguing.csv')
ds.shape
```

```python
ds.columns
```

```python
ds[['Assessments', 'Authority', 'Causation', 'Conditionals',
       'Contrast', 'Difficulty', 'Doubt', 'Emphasis', 'Generalization',
       'Inconsistency', 'Inyourshoes', 'Necessity', 'Possibility', 'Priority',
       'Rhetoricalquestion', 'Structure', 'Wants']].describe()
```

```python
ds['creation_dt'] = pd.to_datetime(ds['timestamp'], unit='s')
ds['creation_date'] = ds.creation_dt.dt.date
```

```python
article = ds[ds.page_title=='Talk:Rafael Correa']
stat = article.groupby('creation_date').agg({'Assessments':'mean'})
stat.head()
```

```python
plt.figure(figsize=(15,5))
#stat.plot.line()
plt.plot(stat)
```
